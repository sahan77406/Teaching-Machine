## **Clean Data**

The objective is to create a model to predict which people would survive based on the following fields:

- **Name** (str) - The passenger name
- **Pclass** (int) - Ticket class
- **Sex** (str) - Sex of the passenger
- **Age** (float) - Age in years
- **SibSp** (int) - Number of siblings and spouses aboard
- **Parch** (int) - Number of parents and children aboard
- **Ticket** (str) - Ticket number
- **Fare** (float) - Passenger fare
- **Cabin** (str) - Cabin number
- **Embarked** (str) - Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

### **Reading Data From CVS File**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

ship = pd.read_csv("data_1.csv", index_col= False)
ship.head()
```

### **Filling Missing Data** ###

```python
ship.isnull().sum()
```

**Replacing missing age with the mean of the existing ages**

```python
ship['Age'].fillna(ship['Age'].mean(), inplace=True)
ship.head(12)
```

### **Combine SibSp & Parch** ##

```python
for i, col in enumerate(['SibSp', 'Parch']):
    plt.figure(i)
    sns.catplot(x=col, y='Survived', data=ship, kind='point', aspect=2 )
```

```python
ship['Family_memb'] = ship['Parch'] + ship ['SibSp']
ship.head()
```

### **Drop Unnecessary Variables** 

```python
ship.drop(['PassengerId', 'SibSp', 'Parch'], axis=1, inplace=True)

ship.head()
```

### **Clean Categorical Variables**

```python
ship.isnull().sum()
```

**Fill in missing and create indicator for cabin**

```python
ship.groupby(ship['Cabin'].isnull())['Survived'].mean()
```

```python
ship['Cabin_ind'] = np.where(ship['Cabin'].isnull(), 0, 1)
ship.head(10)
```

**Convert sex to numeric**

```python
gender = {'male' : 0, 'female' : 1}
ship['Sex'] = ship['Sex'].map(gender)
ship.head(10)
```

**Drop Cabin, Embarked, Name, Ticket**

```python
ship = pd.read_csv("numeric_gender.csv")
ship.drop(['Name','Ticket', 'Embarked', 'Cabin'], axis=1, inplace=True)
ship.to_csv("cleaned_data.csv")
ship.head(10)
```

## **Split data into train, validation, and test set**

```python
import pandas as pd
from sklearn.model_selection import train_test_split

ship = pd.read_csv("cleaned_data.csv", index_col=False)
ship.head(10)
```

```python
ship = pd.read_csv("cleaned_data.csv")
features = ship.drop('Survived', axis=1)
labels = ship['Survived']

x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.4, random_state=42)
x_val, x_test, y_val, y_test = train_test_split(x_test, y_test, test_size=0.5, random_state=42)

data = [y_train, y_test, y_val]

for dataset in (data):
    print ( round(len(dataset)/len(labels), 2))
```

## ***Write out all data**

```python
# in order to "fit" your model, you need to pass in train_features and train_labels.

x_train.pop(x_train.columns[0]) # Remove Unnamed column
x_train.to_csv("train_features.csv", index=False)
y_train.to_csv("train_labels.csv", index=False)


# Once you a that stored model, , you need to pass in test_features, val_features and test_labels and val_labels into predict, in order to generate new "predict"
x_val.pop(x_val.columns[0])  # Remove Unnamed column
x_val.to_csv("val_features.csv", index=False)

x_test.pop(x_test.columns[0])  # Remove Unnamed column
x_test.to_csv("test_features.csv", index=False)

y_val.to_csv("val_labels.csv", index=False)
y_test.to_csv("test_labels.csv", index=False)
```

## Logistic Regression: Fit and evaluate a model

```python
import joblib
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=DeprecationWarning)

tr_features = pd.read_csv('train_features.csv')
tr_labels = pd.read_csv('train_labels.csv', header=None)

tr_labels.to_csv("tr_labels.csv", index=False, header=False)
tr_features.to_csv("tr_features.csv", index=False)
```

### Hyperparameter tuning

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

```python


lr = LogisticRegression()
parameters = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]
}

cv = GridSearchCV(lr, parameters, cv=5)
cv.fit(tr_features, tr_labels.values.ravel())

print_results(cv)
```
